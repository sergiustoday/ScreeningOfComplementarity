```python
from sklearn.datasets import load_boston
import pandas as pd
from sklearn.tree import DecisionTreeRegressor

boston = load_boston()

bos = pd.DataFrame(boston.data)
bos.columns = boston.feature_names
bos['PRICE'] = boston.target

pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)

bos_small = bos.loc[(bos['RM'] >= 5) & (bos['RM'] < 6.7)]
# bos_small = bos

print(bos_small.shape[0])

bos_small
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from scipy.stats import stats as st
from scipy.stats import t
import numpy as np

p_value = 0.05

df_threesome_with_cv = df_threesome_BOS_one

depth = 3
num_items = 100
m = 200

df_threesome_GB = pd.DataFrame(columns = ['f1', 'f2', 'f_screen', 'num_of_rows', \
                                               'screen_coef_mean' , 'screen_coef_lower', 'screen_coef_upper', \
                                               'score_screen_mean', 'score_screen_lower', 'score_screen_upper', \
                                               'score_elim_mean', 'score_elim_lower', 'score_elim_upper', \
                                               ])

for i in range(df_threesome_with_cv.shape[0]):

    scr1_list = []
    scr2_list = []
    for j in range(m):    
        train, test = train_test_split(bos_small, random_state = j)

        nX = df_threesome_with_cv.loc[i, 'f1']
        nY = df_threesome_with_cv.loc[i, 'f2']
        nE = df_threesome_with_cv.loc[i, 'f_screen']

        X_train = train.iloc[:, [nX, nY, nE]].values
        y_train = train['PRICE'].values

        X_test = test.iloc[:, [nX, nY, nE]].values
        y_test = test['PRICE'].values

        model = GradientBoostingRegressor(max_depth = depth, n_estimators = num_items, random_state = 0)
        model.fit(X_train, y_train)
        scr1 = model.score(X_test, y_test)
        scr1_list.append(scr1)

        X_train = train.iloc[:, [nX, nY]].values
        y_train = train['PRICE'].values

        X_test = test.iloc[:, [nX, nY]].values
        y_test = test['PRICE'].values

        model = GradientBoostingRegressor(max_depth = depth, n_estimators = num_items, random_state = 0)
        model.fit(X_train, y_train)
        scr2 = model.score(X_test, y_test)
        scr2_list.append(scr2)

    print(t.interval(alpha = 1-p_value, df = len(scr1_list) - 1, loc = np.mean(scr1_list), scale = st.sem(scr1_list)))
    print(t.interval(alpha = 1-p_value, df = len(scr2_list) - 1, loc = np.mean(scr2_list), scale = st.sem(scr2_list)))

    lower_cutoff1, upper_cutoff1 = \
        t.interval(alpha = 1 - p_value, df = len(scr1_list) - 1, loc = np.mean(scr1_list), scale = st.sem(scr1_list))

    lower_cutoff2, upper_cutoff2 = \
        t.interval(alpha = 1 - p_value, df = len(scr2_list) - 1, loc = np.mean(scr2_list), scale = st.sem(scr2_list))

    print('({0}, {1})'.format(lower_cutoff2 / upper_cutoff1, \
                              upper_cutoff2 / lower_cutoff1))
    print((lower_cutoff2 / upper_cutoff1 + upper_cutoff2 / lower_cutoff1) / 2)
    print('---------------------------------------------------------------------')

    if lower_cutoff2 / upper_cutoff1 > 1.000000001:
        df_threesome_GB = \
        df_threesome_GB.append({'f1': nX, 'f2': nY, 'f_screen': nE, 'num_of_rows': bos_small.shape[0], \
                                     'score_screen_mean': np.mean(scr1_list), \
                                     'score_screen_lower': lower_cutoff1, \
                                     'score_screen_upper': upper_cutoff1, \
                                     'score_elim_mean': np.mean(scr2_list), \
                                     'score_elim_lower': lower_cutoff2, \
                                     'score_elim_upper': upper_cutoff2, \
                                     'screen_coef_mean': (lower_cutoff2/upper_cutoff1+upper_cutoff2/lower_cutoff1)/2,\
                                     'screen_coef_lower': lower_cutoff2 / upper_cutoff1, \
                                     'screen_coef_upper': upper_cutoff2 / lower_cutoff1 \
                                    }, ignore_index = True)
    
df_threesome_GB
```
