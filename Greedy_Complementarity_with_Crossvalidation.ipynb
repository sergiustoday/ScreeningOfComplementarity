```python
from sklearn.datasets import load_boston
import pandas as pd
from sklearn.tree import DecisionTreeRegressor

boston = load_boston()

bos = pd.DataFrame(boston.data)
bos.columns = boston.feature_names
bos['PRICE'] = boston.target

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

bos_small = bos.loc[(bos['RM'] >= 5) & (bos['RM'] < 6.7)]
# bos_small = bos

print(bos_small.shape[0])

bos_small
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
import operator as op
from functools import reduce

def ncr(n, r):
    r = min(r, n-r)
    numer = reduce(op.mul, range(n, n-r, -1), 1)
    denom = reduce(op.mul, range(1, r+1), 1)
    return numer // denom  # or / in Python 2

#    ?
df = bos_small

df_comp = pd.DataFrame(columns = ['f1', 'f2', 'scr_f1', 'scr_f2', 'scr_two', 'root_id'])

num_columns = len(df.columns) - 1
for i in range(num_columns):
    for j in range(i + 1, num_columns):
        
        X = df.iloc[:, i].values.reshape(-1, 1)
        y = df.loc[:, 'PRICE'].values

        model = DecisionTreeRegressor(max_depth = 2, random_state = 0).fit(X, y)
        scr_f1 = model.score(X, y)
        
        X = df.iloc[:, j].values.reshape(-1, 1)
        y = df.loc[:, 'PRICE'].values

        model = DecisionTreeRegressor(max_depth = 2, random_state = 0).fit(X, y)
        scr_f2 = model.score(X, y)
        
        X = df.iloc[:, [i, j]].values
        y = df.loc[:, 'PRICE'].values

        model = DecisionTreeRegressor(max_depth = 2, random_state = 0).fit(X, y)
        scr_two = model.score(X, y)
        
        root_id = model.tree_.feature[0]
        
        df_comp = df_comp.append({'f1': i, 'f2': j, 'scr_f1': scr_f1, 'scr_f2': scr_f2, 'scr_two': scr_two, \
                                  'root_id': root_id}, ignore_index = True)

for i in range(df_comp.shape[0]):
    df_comp.loc[i, 'complementary_force'] = \
        df_comp.loc[i, 'scr_two'] / max(df_comp.loc[i, 'scr_f1'], df_comp.loc[i, 'scr_f2'])
    
print(' ')
print(df_comp)

df_threesome = pd.DataFrame(columns = ['f1', 'f2', 'f3', 'scr_three'])

num_excluded = 0
num_columns = len(df.columns) - 1
for i in range(num_columns):
    for j in range(i + 1, num_columns):
        for k in range(j + 1, num_columns):
            
            #       ?
            num_comp = 0
            num_comp2 = 0
            
            #  i  j
            df_pairs_in_three = df_comp[(df_comp['f1'] == i) & (df_comp['f2'] == j)]
            df_pairs_in_three = df_pairs_in_three.reset_index(drop = True)
            if df_pairs_in_three.loc[0, 'complementary_force'] > 1.00000001:
                num_comp += 1
                
            if df_pairs_in_three.loc[0, 'root_id'] == 0:
                scr_root = 'scr_f1'
            else:
                scr_root = 'scr_f2'
                
            if (df_pairs_in_three.loc[0, 'scr_two'] / df_pairs_in_three.loc[0, scr_root]) > 1.00000001:
                num_comp2 += 1
                
            #  i  k
            df_pairs_in_three = df_comp[(df_comp['f1'] == i) & (df_comp['f2'] == k)]
            df_pairs_in_three = df_pairs_in_three.reset_index(drop = True)
            if df_pairs_in_three.loc[0, 'complementary_force'] > 1.00000001:
                num_comp += 1
                
            if df_pairs_in_three.loc[0, 'root_id'] == 0:
                scr_root = 'scr_f1'
            else:
                scr_root = 'scr_f2'
                
            if (df_pairs_in_three.loc[0, 'scr_two'] / df_pairs_in_three.loc[0, scr_root]) > 1.00000001:
                num_comp2 += 1
                
            #  j  k
            df_pairs_in_three = df_comp[(df_comp['f1'] == j) & (df_comp['f2'] == k)]
            df_pairs_in_three = df_pairs_in_three.reset_index(drop = True)
            if df_pairs_in_three.loc[0, 'complementary_force'] > 1.00000001:
                num_comp += 1
                
            if df_pairs_in_three.loc[0, 'root_id'] == 0:
                scr_root = 'scr_f1'
            else:
                scr_root = 'scr_f2'
                
            if (df_pairs_in_three.loc[0, 'scr_two'] / df_pairs_in_three.loc[0, scr_root]) > 1.00000001:
                num_comp2 += 1
                
                
            if num_comp2 < 2:
                print ('   ')
                print([i, j, k])
                num_excluded += 1
            elif num_comp < 2:
                print ('   ,   ,    !')
                print([i, j, k])
        
            X = df.iloc[:, [i, j, k]].values
            y = df.loc[:, 'PRICE'].values

            model = DecisionTreeRegressor(max_depth = 2, random_state = 0).fit(X, y)
            scr_three = model.score(X, y)

            df_threesome = df_threesome.append({'f1': i, 'f2': j, 'f3': k, 'scr_three': scr_three}, \
                                               ignore_index = True)
            
print(' ')
print(df_threesome)
        
#    df_comp  df_threesome    

#   
df1 = df_comp.loc[:, ['f1', 'scr_f1']]
df1.columns = ['f', 'scr']

df2 = df_comp.loc[:, ['f2', 'scr_f2']]
df2.columns = ['f', 'scr']

df3 = df1.append(df2, ignore_index = True)
df3 = df3.drop_duplicates()
df3['num'] = 1

#   
df4 = df_comp.loc[:, ['f1', 'f2', 'scr_two']]

df5 = pd.DataFrame(columns = ['f', 'scr', 'num']) 
for i in range(df4.shape[0]):
    f1 = df4.loc[i, 'f1']
    f2 = df4.loc[i, 'f2']
    scr_two = df4.loc[i, 'scr_two']
    df5 = df5.append({'f': [f1, f2], 'scr': scr_two, 'num': 2}, ignore_index = True)
    
df6 = df3.append(df5, ignore_index = True)

#   
df7 = pd.DataFrame(columns = ['f', 'scr', 'num']) 
for i in range(df_threesome.shape[0]):
    f1 = df_threesome.loc[i, 'f1']
    f2 = df_threesome.loc[i, 'f2']
    f3 = df_threesome.loc[i, 'f3']
    scr_three = df_threesome.loc[i, 'scr_three']
    df7 = df7.append({'f': [f1, f2, f3], 'scr': scr_three, 'num': 3}, ignore_index = True)
    
df8 = df6.append(df7, ignore_index = True)

df_max_scr = df8[df8['scr'] > (df8['scr'].max()*0.9999999)]

#  - 
df_max_scr_min_num = df_max_scr[df_max_scr['num'] == df_max_scr['num'].min()]
df_max_scr_min_num = df_max_scr_min_num.reset_index(drop = True)

print(' ')
print(df_max_scr_min_num)
print('')

print('   ')
X = df.iloc[:, 0:(df.shape[1] - 1)].values
y = df.loc[:, 'PRICE'].values

model = DecisionTreeRegressor(max_depth = 2, random_state = 0).fit(X, y)
scr_greedy = model.score(X, y)

print(scr_greedy)
print('')

# fn = ['f1', 'f2', 'f3']
tree.plot_tree(model)

#    
koef_screening = df_max_scr_min_num.loc[0, 'scr'] / scr_greedy
print(' : ' + str(koef_screening))

#  
num_max_trees = num_columns + ncr(num_columns, 2) + ncr(num_columns, 3)
koef_speed = (num_max_trees - num_excluded) / num_max_trees
print(' : ' + str(koef_speed))
```

```python
#     

from sklearn.model_selection import train_test_split
import numpy as np
from scipy.stats import stats as st
from scipy.stats import t

# ----------------------------------------------------------------------
# bos_small = bos.loc[(bos['RM'] >= 5) & (bos['RM'] < 6.7)]
# i = 2:  {0.0, 9.0, 3.0} 2.4407537491176243  2*0.45281129546319066
# i = 11: {2.0, 5.0, 6.0} 1.6522268481391817  2*0.2272838475615967
# i = 12: {9.0, 3.0, 6.0} 1.260690176847412   2*0.10819198433486393
# i = 14: {9.0, 5.0, 6.0} 1.6102669630120943  2*0.1921353828985585
# ----------------------------------------------------------------------

p_value = 0.05
n = 1000

# scr1_list = []
# scr2_list = []
df_threesome_with_cv = pd.DataFrame(columns = ['f1', 'f2', 'f_screen', 'num_of_rows', \
                                               'screen_coef_mean' , 'screen_coef_lower', 'screen_coef_upper', \
                                               'score_screen_mean', 'score_screen_lower', 'score_screen_upper', \
                                               'score_elim_mean', 'score_elim_lower', 'score_elim_upper', \
                                               ])

for i in range(df_threesome3.shape[0]):
    scr1_list = []
    scr2_list = []
    
    for j in range(n):    
        train, test = train_test_split(bos_small, random_state = j)

        f1 = df_threesome3.loc[i, 'f1']
        f2 = df_threesome3.loc[i, 'f2']
        f3 = df_threesome3.loc[i, 'f3']
        fe = df_threesome3.loc[i, 'screen_feature']

        fs = [f1, f2, f3]
        fs.remove(fe)

        nX = fs[0]
        nY = fs[1]
        nE = fe

        X_train = train.iloc[:, [nX, nY, nE]].values
        y_train = train['PRICE'].values

        X_test = test.iloc[:, [nX, nY, nE]].values
        y_test = test['PRICE'].values

        model = DecisionTreeRegressor(max_depth = 2, random_state = 0)
        model.fit(X_train, y_train)
        scr1 = model.score(X_test, y_test)
        scr1_list.append(scr1)

        X_train = train.iloc[:, [nX, nY]].values
        y_train = train['PRICE'].values

        X_test = test.iloc[:, [nX, nY]].values
        y_test = test['PRICE'].values

        model = DecisionTreeRegressor(max_depth = 2, random_state = 0)
        model.fit(X_train, y_train)
        scr2 = model.score(X_test, y_test)
        scr2_list.append(scr2)

    print(t.interval(alpha = 1-p_value, df = len(scr1_list) - 1, loc = np.mean(scr1_list), scale = st.sem(scr1_list)))
    print(t.interval(alpha = 1-p_value, df = len(scr2_list) - 1, loc = np.mean(scr2_list), scale = st.sem(scr2_list)))

    lower_cutoff1, upper_cutoff1 = \
        t.interval(alpha = 1 - p_value, df = len(scr1_list) - 1, loc = np.mean(scr1_list), scale = st.sem(scr1_list))

    lower_cutoff2, upper_cutoff2 = \
        t.interval(alpha = 1 - p_value, df = len(scr2_list) - 1, loc = np.mean(scr2_list), scale = st.sem(scr2_list))

    print('({0}, {1})'.format(lower_cutoff2 / upper_cutoff1, \
                              upper_cutoff2 / lower_cutoff1))
    print((lower_cutoff2 / upper_cutoff1 + upper_cutoff2 / lower_cutoff1) / 2)
    print('---------------------------------------------------------------------')

    if lower_cutoff2 / upper_cutoff1 > 1.000000001:
        df_threesome_with_cv = \
        df_threesome_with_cv.append({'f1': nX, 'f2': nY, 'f_screen': nE, 'num_of_rows': bos_small.shape[0], \
                                     'score_screen_mean': np.mean(scr1_list), \
                                     'score_screen_lower': lower_cutoff1, \
                                     'score_screen_upper': upper_cutoff1, \
                                     'score_elim_mean': np.mean(scr2_list), \
                                     'score_elim_lower': lower_cutoff2, \
                                     'score_elim_upper': upper_cutoff2, \
                                     'screen_coef_mean': (lower_cutoff2/upper_cutoff1+upper_cutoff2/lower_cutoff1)/2,\
                                     'screen_coef_lower': lower_cutoff2 / upper_cutoff1, \
                                     'screen_coef_upper': upper_cutoff2 / lower_cutoff1 \
                                    }, ignore_index = True)
    
df_threesome_with_cv
```
